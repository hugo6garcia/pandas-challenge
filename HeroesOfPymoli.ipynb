```python
import pandas as pd
import numpy as np
csv_path = "purchase_data.csv"
game_df = pd.read_csv(csv_path)
game_df.head()
```

```python
player_count = len(game_df["SN"].value_counts())
count

total_count = pd.DataFrame({"Total Players":[player_count]})
total_count
```

```python
unique_items = len(game_df["Item ID"].value_counts())
count
average = game_df["Price"].mean()
average
purchases = game_df["Item ID"].count()
purchases
total = game_df["Price"].sum()
total
summary_df = pd.DataFrame({"Unique Items": [unique_items],
                           "Average Price": [average],
                           "Number of Purchases": [purchases],
                           "Total Revenue": [total]})


summary_df.head()
```

```python
gender = game_df.groupby("Gender")

total_gender = gender.nunique()["SN"]

percentage_of_players = total_gender / player_count *100

gender_unique = pd.DataFrame({"Percentage of Players": percentage_of_players,
                              "Total Count": total_gender})

gender_unique.index.name = None

gender_unique.sort_values(["Total Count"], ascending = False).style.format({"Percentage of Players":"{:.2f}"})

gender_unique.head()
```

```python
purchase_count = gender["Purchase ID"].count()

avg_purchase_price = gender["Price"].mean()

avg_purchase_total = gender["Price"].sum()

avg_purchase_total_person = avg_purchase_total / total_gender

gender_unique = pd.DataFrame({"Purchase Count": purchase_count,
                              "Average Purchase Price": avg_purchase_price,
                             "Average Purchase Total": avg_purchase_total,
                             "Average Purchase Per Person": avg_purchase_total_person})

gender_unique.head()
```

```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

game_df["Age Group"] = pd.cut(game_df["Age"], age_bins, labels=group_names)
game_df


group_age = game_df.groupby("Age Group")

total_count_age = group_age["SN"].nunique()

percentage_age = (total_count_age/ player_count) *100

age_data = pd.DataFrame({"Percentage of Players": percentage_age, "Total Count": total_count_age})

age_data.index.name = None

age_data.head(10)

```

```python
purchase_count_age = group_age["Purchase ID"].count()

avg_purchase_price_age = group_age["Price"].mean()

total_purchase = group_age["Price"].sum()

avg_purchase_per_person_age = total_purchase /total_count_age

age_data = pd.DataFrame({"Purchase Count": purchase_count_age,
                                 "Average Purchase Price": avg_purchase_price_age,
                                 "Total Purchase Value":total_purchase,
                                 "Average Purchase Total per Person": avg_purchase_per_person_age})

age_data.index.name = None

age_data.head()
```

```python
spender_data = game_df.groupby("SN")

purchase_count_spender = spender_data["Purchase ID"].count()

avg_purchase_price_spender = spender_data["Price"].mean()

purchase_total_spender = spender_data["Price"].sum()

spender = pd.DataFrame({"Purchase Count": purchase_count_spender,
                             "Average Purchase Price": avg_purchase_price_spender,
                             "Total Purchase Value":purchase_total_spender})

format_spenders = spender.sort_values(["Total Purchase Value"], ascending=False).head()

format_spenders.head()

```

```python
items = game_df[["Item ID", "Item Name", "Price"]]

item_data = items.groupby(["Item ID","Item Name"])

purchase_count = item_data["Price"].count()
 
purchase_total = (item_data["Price"].sum()) 

item_price = purchase_total/purchase_count

popular_items = pd.DataFrame({"Purchase Count": purchase_count, 
                                "Item Price": item_price,
                                "Total Purchase Value":purchase_total})

popular_format = popular_items.sort_values(["Purchase Count"], ascending=False).head()

popular_format.head()
```
